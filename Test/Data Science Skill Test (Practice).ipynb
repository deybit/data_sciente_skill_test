## Creamos una estructura multi-index para separar los valores de correlacion para SalePrice

```python
#Creamos una estructura multi-index para separar los valores de correlacion para SalePrice
corr_unstack = corr_df.unstack()
print (corr_unstack)
sale_corr = corr_unstack.loc['SalePrice']
```

## Con esto nos quedamos con aquellas nombres de columnas que sean relevantes, tomando aquellas cuya correlación sea superior a 0.3, lo ideal es a partir de 0.5 pero por ser un dataset pequeño no podemos deshacernos de mucha cantidad de datos

```python
#Con esto nos quedamos con aquellas nombres de columnas que sean relevantes
sale_corr_2 = sale_corr[sale_corr < 0.3]
remove_list = list(sale_corr_2.index)
df_3 = df_3.drop(remove_list, axis=1)
```

## Preparamos los datos para el entrenamiento

```python
X = df_3.drop(['SalePrice'], axis=1) #Datos para la prueba

Y = df_3.iloc[:,-1] #Columna objetivo SalePrice

X_train, X_test, y_train, y_test = tts(X, Y, test_size=0.35, random_state=40)
```

## Hacemos uno de una regresión lineal como modelo

```python
lr_2 = LinearRegression()
lr_2.fit(X_train, y_train)
```

## Evaluamos el modelo de regresión lineal entrenado

```python
y_lr_train_pred = lr_2.predict(X_train)
y_lr_test_pred = lr_2.predict(X_test)

lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred)
lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)

lr_results = pd.DataFrame(['Linear regression',lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
lr_results.columns = ['Method','Training MSE','Training R2','Test MSE','Test R2']

lr_results
```

## De acuerdo con https://scikit-learn.org/stable/modules/cross_validation.html#cross-validation haciendo una validación cruzada, tenemos que el desempeño del estimador es de 79.785%

## Plot del grafico del entrenamiento 

```python
#Plot del grafico del entrenamiento 
plt.figure(figsize=(20,20))
plt.scatter(x=y_train, y=y_lr_train_pred, c="#1111FF", alpha=0.3)
z = np.polyfit(y_train, y_lr_train_pred, 1)
p = np.poly1d(z)
plt.plot(y_train,p(y_train),"#000000")
plt.ylabel('Prediccion de  SalesPrice')
plt.xlabel('Valor de SalesPrice')
```
